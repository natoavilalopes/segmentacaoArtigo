```python
def load_classes():
    #with open('/data.yaml', 'r') as config_file:
    with open('/content/segmentacao/model/data.yaml', 'r') as config_file:
        config_data = config_file.read()

    class_names_start = config_data.find("names: [") + len("names: [")
    class_names_end = config_data.find("]", class_names_start)
    class_names_str = config_data[class_names_start:class_names_end]
    class_names = [name.strip().strip("'") for name in class_names_str.split(",")]

    return class_names

nomeClasses = load_classes()
```

```python
import random

def class_colors(names):
    return {name: (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)) for name in names}

color = class_colors(nomeClasses)
```

```python
import cv2
from ultralytics import YOLO

fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_filename = '/content/segmentacao/out/mascara.avi'
frame_rate = 10.0

classes = load_classes()
color = class_colors(classes)

model = YOLO("/content/segmentacao/model/best.pt")

video = '/content/segmentacao/videos/estrada05.mp4'

cap = cv2.VideoCapture(video)
```

```python
import numpy as np

definido = 0

while True:

    ret, frame = cap.read()
    if not ret:
        print("Falha ao ler o vídeo.")
        break

    size = (frame.shape[1], frame.shape[0])
    min_y = int(frame.shape[0] * 0.25)
    mask_height = frame.shape[0] - min_y

    results = model(frame)
    result = results[0]

    segmentation_contours_idx = []
    for seg in result.masks.xyn:
        seg[:, 0] *= size[0]
        seg[:, 1] *= size[1]
        segment = np.array(seg, dtype=np.int32)
        segmentation_contours_idx.append(segment)

    bboxes = np.array(result.boxes.xyxy.cpu(), dtype="int")
    yoloClasses = np.array(result.boxes.cls.cpu(), dtype="int")
    scores = np.array(result.boxes.conf.cpu())

    mask = np.zeros((mask_height, frame.shape[1]), dtype=np.uint8)

    for bbox, class_id, seg, score in zip(bboxes, yoloClasses, segmentation_contours_idx, scores):
        className = classes[class_id]

        if className == "rural_road" or className == "road":
            seg = seg[seg[:, 1] >= min_y]
            seg[:, 1] -= min_y
            cv2.fillPoly(mask, [seg], 255)

        cv2.polylines(frame, [seg], True, color[className], 3)

    mask_stretched = np.zeros((mask_height * 3, frame.shape[1]), dtype=np.uint8)

    for y in range(mask_height):
        mask_stretched[3 * y:3 * (y + 1), :] = mask[y, :]

    if definido == 0:
        sizeMask = (mask_stretched.shape[1], mask_stretched.shape[0])
        video_writer = cv2.VideoWriter(output_filename, fourcc, frame_rate, sizeMask, isColor=False)
        if not video_writer.isOpened():
            print("Erro ao abrir o video_writer. Verifique o codec e o tamanho.")
            break
        definido = 1

    video_writer.write(mask_stretched)

    #vid = cv2.resize(mask_stretched, None, fx=0.4, fy=0.4)
    #cv2.imshow("Image", vid)
    #if cv2.waitKey(1) & 0xFF == ord('q'):
    #    break
```

```python
from IPython.display import HTML
from base64 import b64encode

def show_video(video_path, video_width = 640):

  video_file = open(video_path, "r+b").read()

  video_url = f"data:video/mp4;base64,{b64encode(video_file).decode()}"
  return HTML(f"""<video width={video_width} controls><source src="{video_url}"></video>""")
```
